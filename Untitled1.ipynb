```python

def plot_images(np_images, titles = ['Volume', 'Ground Truth', 'Predicted', 'Uncertainty Map'], columns = 5, figure_size = (24, 18)):
    count = np_images.shape[0]
    rows = math.ceil(count / columns)

    fig = plt.figure(figsize=figure_size)
    subplots = []
    for index in range(count):
        ax = fig.add_subplot(rows, columns, index + 1)
        ax.set_yticklabels([])
        ax.set_xticklabels([])
        subplots.append(ax)
        if len(titles) and index < 4:
            subplots[-1].set_title(str(titles[index]))
        cmp = 'gray'
        if index %4==0:
            cmp = 'gray'
        else:
            np_images[index] = 1-np_images[index]
        plt.imshow(np_images[index], cmap=cmp)
        plt.imsave(f'{index}.png', np_images[index], cmap=cmp)
#         CS3 = plt.contourf(, levels, cmap=cmp)
#         if index %4 == 0:
#             fig.colorbar(cm.ScalarMappable(norm=norm, cmap=cmap), ax=ax)
    
    plt.show()
```
