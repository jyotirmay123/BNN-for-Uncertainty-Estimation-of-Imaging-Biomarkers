```python
import matplotlib.pyplot as plt
import nibabel as nb
import numpy as np
import glob
%matplotlib inline
```

```python
model = 'punet_v4_1'
dataset = 'UKB'
vol = "1005379_20201_2_0"
```

```python
data_path = '/home/abhijit/Jyotirmay/thesis/hquicknat/datasets/'+dataset+'/processed_data/data/'+vol+'.nii.gz'
label_path = '/home/abhijit/Jyotirmay/thesis/hquicknat/datasets/'+dataset+'/processed_data/labels/'+vol+'.nii.gz'
seg_path = '/home/abhijit/Jyotirmay/thesis/hquicknat/outs/'+model+'/'+model+'_predictions_'+dataset+'/'+vol+'_seg.nii.gz'
uncertainty_path = '/home/abhijit/Jyotirmay/thesis/hquicknat/outs/'+model+'/'+model+'_predictions_'+dataset+'/'+vol+'_uncertainty.nii.gz'

sample_paths = '/home/abhijit/Jyotirmay/thesis/hquicknat/outs/'+model+'/'+model+'_predictions_'+dataset+'/'+vol+'_samples/*'
```

```python
data_ = nb.load(data_path)
data = data_.get_fdata()

label_ = nb.load(label_path)
label = label_.get_fdata()

seg_ = nb.load(seg_path)
seg = seg_.get_fdata()

uncertainty_ = nb.load(uncertainty_path)
uncertainty = uncertainty_.get_fdata()

samples = []
for sample in glob.glob(sample_paths):
    s = nb.load(sample)
    samples.append(s.get_fdata())
```

```python
ncols = 14
nrows = 10
s = 10
fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 40))
start_from = 10

for i in range(nrows):
    n = start_from + i
    ax[i][0].imshow(np.squeeze(data[n]), cmap='gray', vmin=0, vmax=2)
    ax[i][0].set_title("Input Image", fontsize=10, color="blue")
    ax[i][0].axis('off')
    ax[i][1].imshow(label[n], cmap='CMRmap', vmin=0, vmax=2)
    ax[i][1].set_title("Ground Truth", fontsize=10, color="blue")
    ax[i][1].axis('off')
    ax[i][2].imshow(seg[n], cmap='CMRmap', vmin=0, vmax=2)
    ax[i][2].set_title("Predicted", fontsize=10, color="blue")
    ax[i][2].axis('off')
    ax[i][3].imshow(uncertainty[n], cmap='gray', vmin=0, vmax=2)
    ax[i][3].set_title("Uncertainty", fontsize=10, color="blue")
    ax[i][3].axis('off')
    for j in range(s):
        k = j+4
        ax[i][k].imshow(samples[j][n], cmap='CMRmap', vmin=0, vmax=2)
        ax[i][k].set_title("Sample", fontsize=10, color="blue")
        ax[i][k].axis('off')
```
