```python

import matplotlib.pyplot as plt
%matplotlib inline

import os
from glob import glob
import nrrd #pip install pynrrd, if pynrrd is not already installed
import nibabel as nib #pip install nibabel, if nibabel is not already installed
import numpy as np
```

```python
kora_set_dir = '/home/abhijit/nas_drive/Data_WholeBody/KORA/KORA_all/KORA_Nifti/*' 
kora_set_paths = glob(kora_set_dir)
print(kora_set_paths)
```

```python
corrupted_paths = []
for kora_path in kora_set_paths:
    kora_opp_files = glob(kora_path+'/*opp_*.nii.gz')
    try:
        upperFile = glob(kora_path+'/*opp_8.nii.gz')[0]
        lowerFile = glob(kora_path+'/*opp_9.nii.gz')[0]
        u = nib.load(upperFile)
        ud = u.get_fdata()
        
        l = nib.load(lowerFile)
        ld = l.get_fdata()
        
        header = l.header
        
        print(ud.shape, ld.shape)
        out = np.hstack((ud, ld))
        mgz = nib.MGHImage(np.squeeze(out), np.eye(4), header=header)
        nib.save(mgz, 'expt.mgz')
        break
        
    except Exception as e:
        print(e, kora_path)
        corrupted_paths.append(kora_path)
        continue
    
```
