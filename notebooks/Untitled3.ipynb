```python
for d in [df_h_k, df_h_n, df_h_u, df_f_k, df_f_n, df_f_u, df_m_k, df_m_n, df_m_u, df_p_k, df_p_n, df_p_u]:
    d.loc[0] = None
    d.dropna(inplace=True) 
```
