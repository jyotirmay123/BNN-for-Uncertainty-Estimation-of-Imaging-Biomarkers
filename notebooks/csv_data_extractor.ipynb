```python
import pandas as pd
import numpy as np
import nibabel as nb
import glob
```

```python
def split_list(a_list):
    half = len(a_list)//2
    half = half + 1
    return a_list[:half], a_list[half:]

def print_affix(a_list, afx=''):
    _ = [print(afx+str(l)) for l in a_list]
```

```python
df_f = pd.read_csv('/home/abhijit/Jyotirmay/my_thesis/dataset_groups/whole_body_datasets/KORA/KORA-diabetes-english.csv')
# df_f.head()
# df_f.count()
```

```python
data = glob.glob('/home/abhijit/Jyotirmay/my_thesis/dataset_groups/whole_body_datasets/KORA/processed_data/data/*')
print(len(data), data[2])

data_idxs = [d.split('/')[-1].split('.')[0][4:] for d in data]
print(len(data_idxs), data_idxs[2])                 
```

```python
df = df_f[df_f['mrt-image-id'].isin(data_idxs)]
```

```python
# df.count()
```

```python
categorical_features = ['smoker', 'bmi-who']
for f in categorical_features:
    dfDummies = pd.get_dummies(df[f], prefix = f)
    df = pd.concat([df, dfDummies], axis=1)
    
yes_no_features = ['hypertension', 'meds-antidiabetic', 'meds-antihypertensive', 'meds-incretin-mimetics', 'meds-insulin-therapy', 'meds-lipoprotein-lowering', 'meds-oral-antidiabetic', 'mri-liver-fat-artifacts']
for f in yes_no_features:
    df[f] = df[f].map(dict(yes=1, no=0))
    
df['sex'] = df['sex'].map(dict(male=1, female=0)) 

# # 0 ='normal', 1 = 'obesity class I', 2 = 'obesity class II', 3 = 'obesity class III', 4 = 'pre-obisety'
# df['bmi-who'] = df.groupby(['bmi-who']).ngroup()

# 0 ='normal', 1 = 'pre-diabetes', 2 = 'type-2-diabetes'
df['diabetes_status'] = df.groupby(['diabetes_status']).ngroup()
```

```python
df.to_csv('/home/abhijit/Jyotirmay/my_thesis/dataset_groups/whole_body_datasets/KORA/processsed_csv.csv')
```

```python
normal_ids = df[df['diabetes_status']==0]['mrt-image-id'].values
pre_diabetes_ids = df[df['diabetes_status']==1]['mrt-image-id'].values
diabetes_ids = df[df['diabetes_status']==2]['mrt-image-id'].values

print('normal:', len(normal_ids), 'pre_diabetes:', len(pre_diabetes_ids), 'diabetes:', len(diabetes_ids))

```

```python
n1, n2 = split_list(normal_ids)
pd1, pd2 = split_list(pre_diabetes_ids)
d1, d2 = split_list(diabetes_ids)

print('normal:', len(n1), 'pre_diabetes:', len(pd1), 'diabetes:', len(d1))
print('normal:', len(n2), 'pre_diabetes:', len(pd2), 'diabetes:', len(d2))

to_train = []
to_test = []

to_train.extend(n1)
to_train.extend(pd1)
to_train.extend(d1)

to_test.extend(n2)
to_test.extend(pd2)
to_test.extend(d2)

print('total_train:', len(to_train))
print('total_test:', len(to_test))
```

```python
print_affix(to_train, 'KORA')
```

```python
print_affix(to_test, 'KORA')
```

```python
n21, n22 = split_list(n2)
pd21, pd22 = split_list(pd2)
d21, d22 = split_list(d2)

print('normal:', len(n21), 'pre_diabetes:', len(pd21), 'diabetes:', len(d21))
print('normal:', len(n22), 'pre_diabetes:', len(pd22), 'diabetes:', len(d22))

to_train_linear = []
to_test_linear = []

to_train_linear.extend(n21)
to_train_linear.extend(pd21)
to_train_linear.extend(d21)

to_test_linear.extend(n22)
to_test_linear.extend(pd22)
to_test_linear.extend(d22)

print('total_train:', len(to_train_linear))
print('total_test:', len(to_test_linear))
```
